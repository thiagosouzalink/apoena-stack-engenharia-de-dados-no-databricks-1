```
%sql
CREATE OR REPLACE TEMPORARY VIEW vw_customers(id, name)
AS VALUES
(1, "iury"),
(2, NULL),
(3, "davi"),
(4, NULL),
(5, "davi");
```

```
%sql
SELECT count(*)
FROM vw_customers;
```

```
%sql
SELECT count(name)
FROM vw_customers;
```

```
%sql
SELECT count(DISTINCT name)
FROM vw_customers;
```

```
%sql
SELECT count(ALL name)
FROM vw_customers;
```

```
%sql
SELECT count(id, name)
FROM vw_customers;
```

```
%sql
SELECT count(DISTINCT id, name)
FROM vw_customers;
```

```
%sql
SELECT count(*)
FROM vw_customers
WHERE name IS NOT NULL;
```

```
%sql
SELECT count_if(name IS NOT NULL)
FROM vw_customers;
```

```
%sql
SELECT count_if(name IS NULL)
FROM vw_customers;
```

```
%sql
SELECT count_if(name IS NULL)
filter(WHERE id > 2)
FROM vw_customers;
```

```
%sql
SELECT count_if(name IS NULL AND id > 2)
FROM vw_customers;
```
